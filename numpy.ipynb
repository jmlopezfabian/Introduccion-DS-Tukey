ECM

$$ECM = \frac{1}{n}\sum(y - \hat{y}) ^2$$

```python
y = np.array([1,2,3])
y_hat = np.array([2,4,6])

print(y - y_hat)
print(((y - y_hat) ** 2).sum())

def ecm(y,y_hat):
    """Funcion para calcular el errore cuadratico medio de dos vectores
    
    Se asume que y y y_hat tienen el mismo tamaño
    
    """
    if len(y) != len(y_hat):
        print("Los vectores tienen que tener el mismo tamaño")
    
    if not isinstance(y, np.ndarray) or not isinstance(y_hat, np.ndarray):
        y = np.array(y)
        y_hat = np.array(y_hat)
    
    diferencia_cuadrado = (y - y_hat)**2
    suma = diferencia_cuadrado.sum()
    return suma / len(y)

    return ((y - y_hat) ** 2).sum() / len(y)

ecm([2,4,6],[1,2,3])
```

```python
def pendiente(b: float, x: np.ndarray, y: np.ndarray) -> float:
    return (((x * y).sum()) - b * x.sum()) / (x ** 2).sum()

pendiente(5,x,y)
```
