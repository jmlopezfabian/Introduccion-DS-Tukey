# EDA

```python
categoricas = ['gender', 'course', 'internet_access', 'sleep_quality', 'study_method', 'facility_rating', 'exam_difficulty']
df_muestra = df.sample(1000, replace=False)
for categorica in categoricas:
    sns.scatterplot(x = 'study_hours', y = 'exam_score', data = df_muestra, hue = categorica)
    plt.title(f'Horas de estudio vs calificacion examen (coloreado por {categorica})')
    plt.show()
```

```python
categoricas = ['gender', 'course', 'internet_access', 'sleep_quality', 'study_method', 'facility_rating', 'exam_difficulty']
for categorica in categoricas:
    sns.barplot(data = df, y = 'exam_score', hue=categorica)
    plt.show()
```

```python
df['study_method - sleep_quality'] = df['study_method'] + ' - ' + df['sleep_quality']
df['study_method - sleep_quality']
```

```python
sns.scatterplot(data = df, x='study_hours', y = "exam_score", hue = 'study_method - sleep_quality')
plt.show()
```

```python
plt.figure(figsize=(25,10))
sns.boxplot(data = df.sample(1000, replace = False), y = "exam_score", hue = 'study_method - sleep_quality')
plt.show()
```

```python
df.groupby(by='study_method - sleep_quality')['exam_score'].agg(['mean', 'std','median']).sort_values(by='mean', ascending = False)
```

```python
df_numerico = df.drop(categoricas,axis = 1)
df_numerico.drop('study_method - sleep_quality', axis = 1, inplace = True)
df_numerico.drop('student_id', axis = 1, inplace= True)

matriz = df_numerico.corr() #Matriz de correlacion lineal de pearson
sns.heatmap(matriz, annot = True)
```
