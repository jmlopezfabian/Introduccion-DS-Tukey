```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

plt.style.use("grayscale")
```

```python
#linspaces -> crear un vector [0, 10] espacio en 0.0001

# y = 3x + 10
x = np.linspace(-10,10, 50)
y1 = 3 * x ** 3 + x ** 2 + 10
y2 = np.sin(x) * 1000

plt.plot(x,y1, color = "red", label = "y1", linestyle = "--", marker = "*")
plt.plot(x,y2, color = "blue", label = "y2", linestyle = "--", marker = "^")

plt.title("Graficas y1 y y2")
plt.xlabel("Eje x")
plt.ylabel("Eje y")

plt.legend()

plt.grid(True)
```
