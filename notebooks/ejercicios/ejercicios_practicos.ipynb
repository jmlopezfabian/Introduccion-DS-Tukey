## Fórmulas

Para calcular la regresión lineal simple, necesitas la pendiente ($m$) y la ordenada al origen ($b$):

**Pendiente:**
$$m = \frac{\sum(x_i \cdot y_i) - b \cdot \sum(x_i)}{\sum(x_i^2)}$$

**Ordenada al origen:**
$$b = \bar{y} - m \cdot \bar{x}$$

O usando la fórmula cerrada:

$$m = \frac{n\sum(x_i y_i) - \sum(x_i)\sum(y_i)}{n\sum(x_i^2) - (\sum(x_i))^2}$$

$$b = \frac{\sum(y_i) - m\sum(x_i)}{n}$$

## Este codigo genera ambos conjuntos de datos, no hay que modificarlo

````python
# Generar dos conjuntos de datos aleatorios
# Fijar semilla para reproducibilidad
np.random.seed(42)

# Conjunto de datos 1: RELACIÓN LINEAL
# Generar x con valores aleatorios
n1 = 100
x1 = np.linspace(0, 10, n1)

# Crear relación lineal: y = 2*x + 3 + ruido
m_real_1 = 2.0
b_real_1 = 3.0
ruido_1 = np.random.normal(0, 1.5, n1)  # Ruido gaussiano con media 0 y desviación estándar 1.5
y1 = m_real_1 * x1 + b_real_1 + ruido_1

print("Conjunto 1 (Lineal) generado:")
print(f"  Número de puntos: {n1}")
print(f"  Relación real: y = {m_real_1}*x + {b_real_1} + ruido")
print(f"  Rango de x: [{x1.min():.2f}, {x1.max():.2f}]")
print(f"  Rango de y: [{y1.min():.2f}, {y1.max():.2f}]")

# Conjunto de datos 2: RELACIÓN NO LINEAL (cuadrática)
# Generar x con valores aleatorios
n2 = 100
x2 = np.linspace(0, 10, n2)

# Crear relación no lineal: y = 0.5*x² - 2*x + 5 + ruido
ruido_2 = np.random.normal(0, 2.0, n2)  # Ruido gaussiano con media 0 y desviación estándar 2.0
y2 = 0.5 * x2**2 - 2 * x2 + 5 + ruido_2

print("\nConjunto 2 (No Lineal - cuadrático) generado:")
print(f"  Número de puntos: {n2}")
print(f"  Relación real: y = 0.5*x² - 2*x + 5 + ruido")
print(f"  Rango de x: [{x2.min():.2f}, {x2.max():.2f}]")
print(f"  Rango de y: [{y2.min():.2f}, {y2.max():.2f}]")

# Visualizar los datos originales
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].scatter(x1, y1, alpha=0.6, color='blue')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Conjunto 1: Relación Lineal')
axes[0].grid(True, alpha=0.3)

axes[1].scatter(x2, y2, alpha=0.6, color='red')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Conjunto 2: Relación No Lineal (Cuadrática)')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````
