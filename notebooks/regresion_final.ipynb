```python
var_objetivo = "exam_score"
variables_predicivas = ["study_hours"]

columnas_nuevas = len(df["study_method"].unique()) - 1
nombre_columnas_nuevas = list(df_encoded.columns[-columnas_nuevas:])
variables_predicivas = variables_predicivas + nombre_columnas_nuevas
print(variables_predicivas)

#Para obtener la lista de las variables predictivas, sabiendo que study hours es una de ellas
#y study_mthod tmb pero se agregaron columnas binarias apartir de los valores de study_method

X = df_encoded[variables_predicivas]
y = df_encoded[var_objetivo]
```

# Train / test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

import statsmodels.api as sm

#X_train = sm.add_constant(X_train)

model = sm.OLS(y_train, X_train)

result = model.fit()

result.summary()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

#predicciones para ambos conjunts
predicciones_train = result.predict(X_train)
predicciones_test = result.predict(X_test)



#Train
print(mean_squared_error(y_train, predicciones_train) ** 0.5)

#Test
print(mean_squared_error(y_test, predicciones_test) ** 0.5)
```

# K - FOLD Cross Validation

```python
from sklearn.model_selection import KFold, cross_val_predict
from sklearn.metrics import mean_squared_error

#Hacer que X y Y sean arreglos de numpy
X = np.array(X)
y = np.array(y)


# Configurar el K-Fold
kf = KFold(n_splits=10, shuffle=True, random_state=42)

#Crear listas de las metricas
lista_mse = []
lista_rmse = []
lista_rmse_train = []

# Iteracion sobre los folds

for train_index, test_index in kf.split(X):
    #Dividir X y Y en train/test
    X_train = X[train_index]
    X_test = X[test_index]
    y_train = y[train_index]
    y_test = y[test_index]
    
    #Entrenamiento del modelo
    model = sm.OLS(y_train, X_train).fit()

    #Obtener metricas para el entrenamiento
    predicciones_train = model.predict(X_train)
    rmse_train = mean_squared_error(y_train, predicciones_train) ** 0.5
    lista_rmse_train.append(rmse_train)

    #Obtener metricas para el test
    predicciones = model.predict(X_test)
    mse = mean_squared_error(y_test, predicciones)
    rmse = mse ** 0.5

    #Agregar cada metrica en su lista
    lista_mse.append(mse)
    lista_rmse.append(rmse)

lista_mse = np.array(lista_mse)
lista_rmse = np.array(lista_rmse)
lista_rmse_train = np.array(lista_rmse_train)

print("Media de MSE K - FOLD", lista_mse.mean())
print("Desv Estandar de MSE K - FOLD", lista_mse.std())

print("\n")


print("Media de RMSE K - FOLD", lista_rmse.mean())
print("Desv Estandar de RMSE K - FOLD", lista_rmse.std())


print("\n")


print("Media de RMSE K - FOLD (TRAIN)", lista_rmse_train.mean())
print("Desv Estandar de RMSE K - FOLD (TRAIN)", lista_rmse_train.std())
```

# Leave one out Cross validation (LOOCV)

```python
from sklearn.model_selection import LeaveOneOut

loo = LeaveOneOut()
lista_rmse = []

#Anadir columna de unos para el intercepto/bias/beta 0
#X = sm.add_constant(X)

for train_index, test_index in loo.split(X):
    X_train = X[train_index]
    X_test = X[test_index]
    y_train = y[train_index]
    y_test = y[test_index]
    #Entreamiento

    model = sm.OLS(y_train, X_train).fit()

    #Obtener metrica 
    predicciones = model.predict(X_test)
    rmse = mean_squared_error(y_test, predicciones) ** 0.5
    lista_rmse.append(rmse)

lista_rmse = np.array(lista_rmse)

print("Media de RMSE LOOCV", lista_rmse.mean())
print("Desv Estandar de RMSE LOOCV", lista_rmse.std())
```
