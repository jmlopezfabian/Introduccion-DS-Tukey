# Prueba ANOVA (C vs N)

```python
# 1. generar una lista que contenga las listas de los grupos que queremos comparar
grupos = []
for grupo in df["study_method"].unique():
    filtro = df["study_method"] == grupo
    grupo_act = df[filtro]
    grupos.append(list(grupo_act["exam_score"]))
```

```python
from scipy.stats import f_oneway

f_stat, p_value_anova = f_oneway(*grupos)

if p_value_anova < 0.05:
    print("Se rechaza H0: Los metodos de estudio impactan en el examen score")
else:
    print("No se rechaza H0: Los metodos de estudio NO impactan en el examen score")
```

# Prueba Chi cuadrada (C vs C)

¿Existe asociación entre el genero y el acceso a internet?

$H_0 :$ El género y el acceso a internet son independientes / No exsite asociación entre el genero y el acceso a internet

$H_a :$ El género y el acceso a internet no son independientes / Existe asociación entre el género y el acceso a internet

```python
# 1. Obtener la tabla de contingencias
tabla_contingencias = pd.crosstab(df["gender"], df["study_method"])
tabla_contingencias
```

```python
from scipy.stats import chi2_contingency

chi2_stat, p_value_chi2, _, _ = chi2_contingency(tabla_contingencias)
if p_value_chi2 < 0.05:
    print("Se rechaza H0: Existe asociación entre el genero y el acceso a internet")
else:
    print("No se rechaza H0: NO Existe asociación entre el genero y el acceso a internet")
```
